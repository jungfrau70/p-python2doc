```python
%%writefile captureScreens_jeniffer.py
import os
from bs4 import BeautifulSoup as bs
from datetime import datetime
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.action_chains import ActionChains
from selenium.common.exceptions import NoSuchElementException
from time import sleep

screenshots_path = 'ScreenShots'

def login(driver, url, userid, userid_xpath, passwd, passwd_xpath, login_button_xpath):
    # Grafana 로그인 페이지로 이동
    driver.get(url)
    sleep(3)

    try:
        # 고급 버튼 클릭
        driver.find_element_by_xpath('//*[@id="details-button"]').click()
        sleep(3)
        # 이동 버튼 클릭
        driver.find_element_by_xpath('//*[@id="proceed-link"]').click()
        sleep(3)
    except:
        sleep(3)
    finally:
        print("Step 1")
        # UserID 입력
        username = driver.find_element_by_xpath(userid_xpath)                                            
        username.clear()
        username.send_keys(userid)
        sleep(3)

        print("Step 2")
        # Password 입력
        password = driver.find_element_by_xpath(passwd_xpath)
        password.clear()
        password.send_keys(passwd)
        sleep(3)

        print("Step 3")
        # Login 버튼 클릭
        login = driver.find_element_by_xpath(login_button_xpath).click()

        sleep(3)
        return login

options=Options()
# options.headless = True
driver=webdriver.Chrome(options=options)
# driver.set_window_size(1920,1080) 
driver.set_window_size(1680,1050)
driver.implicitly_wait(30)
driver.maximize_window()

# Login 이동
url = "http://10.11.65.167:7900/dashboard/realtimeAdmin"
userid = 'ccsadmin'
passwd = 'ccsadmin!00'
userid_xpath='//*[@id="app"]/div/div[2]/div[1]/form/div[1]/div/input'
passwd_xpath='//*[@id="app"]/div/div[2]/div[1]/form/div[2]/div/input'
login_button_xpath='//*[@id="app"]/div/div[2]/div[1]/div[3]/div'
ret = login(driver, url, userid, userid_xpath, passwd, passwd_xpath, login_button_xpath)
print(ret)
    
elements = []; filenames = []

elements.append(driver.find_element_by_xpath("//span[text()='TMS']"))
elements.append(driver.find_element_by_xpath("//span[text()='Gen1']"))
elements.append(driver.find_element_by_xpath("//span[text()='Gen2']"))
elements.append(driver.find_element_by_xpath("//span[text()='ETC']"))
elements.append(driver.find_element_by_xpath("//span[text()='센터검색']"))
elements.append(driver.find_element_by_xpath("//span[text()='OLD']"))
# elements.append(driver.find_element_by_xpath("//span[text()='NEW']"))
elements.append(driver.find_element_by_xpath("//span[text()='VCRM']"))

# # 파일 목록 (최근 것만)
filenames.append(screenshots_path + '/TMS_' + datetime.now().strftime('%Y%m%d_%H%M') + '.png')
filenames.append(screenshots_path + '/TMS_Gen1_' + datetime.now().strftime('%Y%m%d_%H%M') + '.png')
filenames.append(screenshots_path + '/TMS_Gen2_' + datetime.now().strftime('%Y%m%d_%H%M') + '.png')   
filenames.append(screenshots_path + '/TMS_ETC_' + datetime.now().strftime('%Y%m%d_%H%M') + '.png')
filenames.append(screenshots_path + '/CENTER_SEARCH_' + datetime.now().strftime('%Y%m%d_%H%M') + '.png')
filenames.append(screenshots_path + '/CENTER_SEARCH_OLD_' + datetime.now().strftime('%Y%m%d_%H%M') + '.png')
# filenames.append(screenshots_path + '/CENTER_SEARCH_NEW_' + datetime.now().strftime('%Y%m%d_%H%M') + '.png')
filenames.append(screenshots_path + '/VCRM_' + datetime.now().strftime('%Y%m%d_%H%M') + '.png')

# driver.execute_script("document.body.style.zoom='95%'")

# 소스 코드 보기
# html = driver.page_source
# soup = bs(html, 'html.parser')
# list = soup.select('.breadcrumb-row-label')
# for here in list:
#     print(here)
#html = bs(driver.page_source, 'html.parser')
#print(html)

# OPEN TREE
def open_tree(driver):
    xpath = '//*[@id="vue_app"]/div/div[1]/div[1]/div[2]/div[1]/div/div/div'
    element = driver.find_element_by_xpath(xpath)
    action = ActionChains(driver)
    action.move_to_element(element).click().perform()
    sleep(3)

# CLOSE TREE
def close_tree(driver):
    xpath = '/html/body/div[3]/div[1]/div[1]/span'
    ret = driver.find_element_by_xpath(xpath).click()
    sleep(3)

def capture_screen(driver):
    for (element, filename) in zip(elements, filenames):
        try:
            # Move to each menu
            open_tree(driver)
            action = ActionChains(driver)
            action.move_to_element(element).click().perform()
            close_tree(driver)
        except NoSuchElementException as e:
            print(e)
        finally:
            # Capture screen
#             driver.execute_script("document.body.style.zoom='95%'")
            sleep(15)
            driver.save_screenshot(filename)

# OPEN TREE
def open_tree(driver):
    xpath = '//*[@id="vue_app"]/div/div[1]/div[1]/div[2]/div[1]/div/div/div'
    element = driver.find_element_by_xpath(xpath)
    action = ActionChains(driver)
    action.move_to_element(element).click().perform()
    sleep(3)

# CLOSE TREE
def close_tree(driver):
    xpath = '/html/body/div[3]/div[1]/div[1]/span'
    ret = driver.find_element_by_xpath(xpath).click()
    sleep(3)


if __name__ == '__main__':
    for (element, filename) in zip(elements, filenames):
        try:
            # Move to each menu
            open_tree(driver)
            action = ActionChains(driver)
            action.move_to_element(element).click().perform()
            close_tree(driver)
        except NoSuchElementException as e:
            print(e)
        finally:
            # Capture screen
#             driver.execute_script("document.body.style.zoom='95%'")
            sleep(15)
            driver.save_screenshot(filename)
```
